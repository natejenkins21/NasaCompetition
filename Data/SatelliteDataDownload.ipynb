```python
import subprocess
import pandas as pd
```

```python
satellite_metadata = pd.read_csv("./pm25_satellite_metadata.csv")
```

### Download the maiac satellite data

```python
split = 'train' # Set to train or test
```

```python
s3_maiac_filenames = satellite_metadata[(satellite_metadata['product']=='maiac') &
                                        (satellite_metadata['split']==split)]['us_url'].to_list()
save_path = f"./maiac_satellite_data/{split}"
print("{} files to download".format(len(s3_maiac_filenames)))
```

```python
for filename in s3_maiac_filenames:
    print(filename)
    subprocess.run(["aws", "s3", "cp", filename, save_path, "--no-sign-request"])
```
